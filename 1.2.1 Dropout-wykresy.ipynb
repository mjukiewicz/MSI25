```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie historii
df_with = pd.read_csv('history_with_dropout.csv')
df_without = pd.read_csv('history_without_dropout.csv')

# Wykresy
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(df_with['accuracy'], label='Trening')
plt.plot(df_with['val_accuracy'], label='Test')
plt.title('Model z Dropout')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(df_without['accuracy'], label='Trening')
plt.plot(df_without['val_accuracy'], label='Test')
plt.title('Model bez Dropout')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
