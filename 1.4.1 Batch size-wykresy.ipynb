```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('mnist_histories.csv')

# Wykres dokładności (train/val) dla różnych batch_size
plt.figure(figsize=(10, 5))
for bsz, part in df.groupby('batch_size'):
        plt.plot(part['epoch'], part['val_accuracy'], linestyle='--', label=f'val ={bsz}')
plt.title('Accuracy vs epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

# Wykres straty (train/val)
plt.figure(figsize=(10, 5))
for bsz, part in df.groupby('batch_size'):
    plt.plot(part['epoch'], part['val_loss'], linestyle='--', label=f'val ={bsz}')
plt.title('Loss vs epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()
```
